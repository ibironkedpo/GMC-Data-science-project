2. Write a python program to reverse names.
E.g aloba - abola, aishat - tahsia



```python
def reverse_name(name):
    return name[::-1]

def main():
    name = input("Enter a name: ")
    reversed_name = reverse_name(name)
    print("Reversed name:", reversed_name)

if __name__ == "__main__":
    main()
```
