```python
for i in df:
    if (df[i].isnull().sum() / len(df)).round(2) * 100 < 30:
        if df[i].dtypes != 'O':
            df[i].fillna(df[i].median(), inplace=True)
        else:
            df[i].fillna(df[i].mode()[0], inplace=True)
    else:
        df.drop(i, axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoded = {}
for i in df:
    encoder = LabelEncoder()
    df[i] = encoder.fit_transform(df[i])
    encoded[i + '_encoder'] = encoder
    # joblib.dump(encoder, open(f'{i}_encoder.pkl', 'wb'))

encoded
```

```python
df.head()
```

```python
df.drop('SMode', axis=1, inplace=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('HasDetections', axis=1)
y = df['HasDetections']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=102)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
dt_model = DecisionTreeRegressor()
dt_model.fit(xtrain,ytrain)
cross_validation = dt_model.predict(xtrain)
score = r2_score(cross_validation, ytrain).round(2)

print(f'The Model is {(r2_score(cross_validation, ytrain).round(2)) * 100}% accurate')
if score > 0.75:
    print('Model did not underfit')
else:
    print('Model is underfitted')
```

```python
prediction = dt_model.predict(xtest)
score = r2_score(prediction, ytest).round(2)

print(f'The Model is {(r2_score(prediction, ytest).round(2)) * 100}% accurate')
if score > 0.75:
    print('Model did not underfit')
else:
    print('Model is underfitted')
```
