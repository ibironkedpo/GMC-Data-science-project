```python
for i in ds:
    if (ds[i].isnull().sum() / len(ds)).round(2) * 100 < 30:
        if ds[i].dtypes != 'O':
            ds[i].fillna(ds[i].median(), inplace = True)
        else:
            ds[i].fillna(ds[i].mode()[0], inplace = True)
    else:
        ds.drop(i, axis = 1, inplace = True)
ds.isnull().sum()
```

```python
ds['Expected Price ($1k)'] = pd.to_numeric(ds['Expected Price ($1k)'], errors = 'coerce')

num = ds.select_dtypes(include = 'number')
cat = ds.select_dtypes(exclude = 'number')

print(f'Numerical Columns')
display(num.head())
print(f'Categorical Columns')
display(cat.head())
```

```python
for ExpectedPrice in ds:
    if (ds[ExpectedPrice].isnull().sum() / len(ds)).round(2) * 100 < 30:
        if ds[ExpectedPrice].dtypes != 'O':
            ds[ExpectedPrice].fillna(ds[ExpectedPrice].median(), inplace = True)
        else:
            ds[ExpectedPrice].fillna(ds[ExpectedPrice].mode()[0], inplace = True)
    else:
        ds.drop(ExpectedPrice, axis = 1, inplace = True)
ds.isnull().sum()
```
