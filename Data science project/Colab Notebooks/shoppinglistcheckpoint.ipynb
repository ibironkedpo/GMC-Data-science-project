```python
shopping_list = []
emp = {}
while True:
    options = input('Enter an option: You can choose to add, remove or view your cart.\nChoose quit to exit\n')
    if options == 'add':
        for item in range(4):
            item_to_add = input('What do you want to add?: ')
            shopping_list.append(item_to_add)
            quantity = int(input('Enter a quantity for your item'))
            emp[item_to_add] = quantity
        print('\n')
    elif options == 'remove':
        item_to_remove = input('What do you want to remove?: ')
        shopping_list.remove(item_to_remove)
        print('\n')
    elif options == 'view':
        print('These are the items you added to your cart')
        for item in shopping_list:
            print(item)
            print('\n')
    elif options == 'quit':
        print('Thank you for shopping with us')
        break
    else:
        print('Please put a valid output')


print('\n')
```
