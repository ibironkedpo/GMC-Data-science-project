```python
import math
```

```python
class Calculator:
    def _init_(self):
        self.operations = {'+': self.add, '-': self.sub, '*': self.multiply, '/': self.divide}

    def add_operation(self, symbol, function):
        self.operations[symbol] = function

    def add(self,num1,num2):
        return num1 + num2

    def sub(self,num1,num2):
        return num1 - num2

    def multiply(self,num1,num2):
        return num1 * num2

    def division(self,num1,num2):
        return num1 / num2

    def exponential(self,num1,num2):
        return num1**num2

    def calculate(self,num1,operation_symbol,num2):
      if operation_symbol not in self.operations:
        print("Invalid operation symbol!")
        raise ValueError("Invalid operation symbol!")
      if not (isinstance(num1, (int, float)) and isinstance(num2, (int, float))):
            print("Invalid input numbers!")
            raise ValueError("Invalid input numbers!")
      return self.operations[operation_symbol](num1, num2)
    def exponential(num1, num2):
      return math.exp(num1 ** num2)

    def square_root(num):
      return math.sqrt(num)

    def logarithm(num, base):
      return math.log(num, base)

# Main program
calculator = Calculator()

# Adding advanced mathematical operations
calculator.add_operation('exp', exponential)
calculator.add_operation('sqrt', square_root)
calculator.add_operation('log', logarithm)

while True:
    try:
        num1 = float(input('Enter the 1st number: '))
        operation_symbol = input("Choose an operation (+, -, *, /, ^ for exponentiation, sqrt for square root, log for logarithm) or 'end' to exit: ")
        if operation_symbol == 'end':
            break
        num2 = float(input('Enter the 2nd number: '))

        result = calculator.calculate(num1, operation_symbol, num2)
        print(f'The result of the {operation_symbol} is {result}')

    except (ValueError, ZeroDivisionError) as e:
      print("Error:", e)
      continue


```
